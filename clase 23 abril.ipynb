# INTERPOLACIÓN POLINOMIAL 
## Dado un conjunto de pares $(x_i,f(x_i))$ encontrar una representación $f(x)= \sum a_n \varphi_n (x)$ en algún sentido optima, osea ¿cuales son los valores de los coeficientes $a_n$ para que la representación de una función como su suma de Taylor se acerque a la realidaaa?

Sea un polinomio

$\begin{equation} 
p(x)=c_0+c_1 x+ ...+ c_{N-1}x^{N-1}
\end{equation}$

Estoy tomando que la función pasa por N puntos tq $\varphi_n=x_n$.
Imagintate que conozco todas las $x_i$ y $f_i$, quiero encontrar los $c_N$ que haga que mi representación $p(x)$ se acerque chido a $f(x)$. 

Puedo conocer las $c$ a partir de la sig notacion matricial, sea $\Phi$ la matriz de las $x$

$\begin{equation}
\begin{pmatrix}
\varphi_0 (x_0) & \varphi_1 (x_0) & ... & \varphi_{N-1} (x_0) \\
\varphi_0 (x_1) & \varphi_1 (x_1) & ... & \varphi_{N-1} (x_1) \\
... & ... & ...& ....\\
\end{pmatrix}
\end{equation}$ 
$\begin{equation}
\begin{pmatrix}
c_0 \\
c_1\\
...\\
\end{pmatrix}
\end{equation}$
= $\begin{equation}
\begin{pmatrix}
f_0 \\
f_1\\
...\\
\end{pmatrix}
\end{equation}$

Simplemente con sacar la inversa ya salen las $c$ es decir $c=\Phi⁻1 F$
