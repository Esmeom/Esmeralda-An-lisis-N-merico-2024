```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt

# Definiendo letra como símbolo matemático
x = sy.symbols('x')

#defino el valor numerico que no quiero que supere el resto
tolerancia= 0.1

# Defino función mera mera
f = sy.sin(x)+sy.cos(x)
ffunc = sy.lambdify(x,f)

for orden in range(1, 20):
    # Serie de Taylor de f(x)
    t = sy.series(f, x, 0, orden)
    tt = t.removeO()
    tfunc = sy.lambdify(x,tt)
    
    # Resto
    resto = np.abs(ffunc(7)-tfunc(7))
    if resto <= tolerancia:
        Orden=orden
        break
        
print("Orden sale", Orden)
print("resto", resto)
```

```python
#Graficar las funciones
X=np.linspace(-10,10)
Yf=ffunc(X)
Yt=tfunc(X)
fig,ax=plt.subplots()
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.plot(X,Yf)
plt.plot(X,Yt)
plt.legend(['$f(x)$','serie de f(x)'])
plt.show()
```

```python

Yerr=np.abs(Yf-Yt)
Yord=X**19
plt.plot(X,Yerr)
plt.plot(X,Yord)
plt.legend(['Error absoluto','O($x^19$)'])
plt.show()
```
