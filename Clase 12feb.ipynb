```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt

#definiendo variables como simbolos
x = sy.symbols('x')

#definimos funciones
f = sy.exp(x)-3*sy.cos(x)

g = sy.series(f,x,0,3)

print(f)
print(g)

#da el orden del error
orden = g.getO()
#remueve el temino que estima la magnitud de error
gx = g.removeO()


print(orden)
print(gx)

#usamos lambdify para convertirlas en funciones
ffunc = sy.lambdify(x,f)
gfunc = sy.lambdify(x,gx)

ffunc(3)
gfunc(3)

#grafiquemos
#np.linspace(0,3) para dividir eje x en intervalos desde 0 a 3
x = np.linspace(0,3)

yf = ffunc(x)
yg = gfunc(x)
yerr = np.abs(yf-yg)
yord = x**3
print('a ver')
print(yerr)
print(yord)


plt.plot(x,yf, label = "a")
plt.plot(x,yg, label = "b")
plt.legend()

plt.plot(x,yerr, label = "error")
plt.plot(x,yord, label = "orden")
plt.legend()

plt.grid(True)
plt.show()
```
