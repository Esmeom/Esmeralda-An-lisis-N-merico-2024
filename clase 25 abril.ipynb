```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función Runge
def f(x):
    return 1/(1+25*x**2)

# Función para calcular los polinomios de Lagrange para un conjunto de puntos (x, y)
def lagrange_polynomials(x_valores, y_valores):
    def L(j):
        def poly(x):
            numerator = 1.0
            denominator = 1.0
            for m in range(len(x_valores)):
                if m != j:
                    numerator *= (x - x_valores[m])
                    denominator *= (x_valores[j] - x_valores[m])
            return numerator / denominator
        return poly

    polinomios = [L(j) for j in range(len(x_valores))]
    return polinomios

# Función para evaluar los polinomios en un conjunto de puntos x
def evaluar_polinomios(polinomios, x):
    y_valores = np.zeros_like(x)
    for polinomio in polinomios:
        y_valores += polinomio(x)
    return y_valores

# Generar valores de x y f(x)
x_valores = np.linspace(-10, 10, 10)
y_valores = f(x_valores)

# Calcular polinomios de Lagrange
polinomios = lagrange_polynomials(x_valores, y_valores)
# Generar valores de x para evaluar los polinomios
x_eval = np.linspace(-10, 10, 1000)
# Evaluar los polinomios en los valores de x
y_eval = evaluar_polinomios(polinomios, x_eval)

# Graficar
plt.plot(x_eval, y_eval, label='Polinomio de Lagrange')
plt.scatter(x_valores, y_valores, color='red', label='Puntos (x, y)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Polinomio de Lagrange para los puntos dados')
plt.legend()
plt.grid(True)
plt.show()
```
